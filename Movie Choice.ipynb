```python
import pandas as pd
imdbrat=r"https://datasets.imdbws.com/title.ratings.tsv.gz"
imdbbas=r"https://datasets.imdbws.com/title.basics.tsv.gz"
```

```python
df1 = pd.read_csv(imdbrat, compression='gzip', sep='\t', header=0, quotechar='"', error_bad_lines=False)
df2 = pd.read_csv(imdbbas, compression='gzip', sep='\t', header=0, quotechar='"', error_bad_lines=False)
```

```python
df2
```

```python
df=df1.merge(df2)
```

```python
df=df.replace(to_replace=r'\N', value=0)
```

```python
df['startYear'] = df['startYear'].astype(int)
```
